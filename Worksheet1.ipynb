```python
# Problem 3
import numpy as np

# Arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Addition
addition = x + y
print("x + y:\n", addition)

# 2. Subtraction
subtraction = x - y
print("x - y:\n", subtraction)

# 3. Multiplication
multiplication = x * 3
multiplication_v = v * 2
print("x * 3:\n", multiplication)
print("v * 2:\n", multiplication_v)

# 4. Square
square_x = x ** 2
print("x squared:\n", square_x)

# 5. Dot products
dot_vw = np.dot(v, w)
print("Dot product between v and w:", dot_vw)

dot_xv = np.dot(x, v)
print("Dot product between x and v:\n", dot_xv)

dot_xy = np.dot(x, y)
print("Dot product between x and y:\n", dot_xy)

# 6. Concatenation
x_with_y = np.vstack((x, y))
print("Vertical stack of x and y:\n", x_with_y)

v_with_w = np.column_stack((v, w))
print("Column stack of v and w:\n", v_with_w)

# 7. Concatenate x with v (reshape v first)
v_reshaped = v.reshape(1, -1)  # makes it 2D
x_with_v = np.vstack((x, v_reshaped))
print("x concatenated with reshaped v:\n", x_with_v)
```

```python
# Problem 4
import numpy as np

# Matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Identity matrix
I = np.eye(2, dtype=int)

# Inverse of A
A_inv = np.linalg.inv(A)

# Multiply A with its inverse
AdotAinv = np.dot(A, A_inv)

# Round the result to nearest integer for comparison
AdotAinv_int = np.round(AdotAinv).astype(int)
print("A * A_inv (rounded to int):\n", AdotAinv_int)

# Check if it's equal to identity matrix
if np.array_equal(AdotAinv_int, I):
    print("True")
else:
    print("False")
```

```python
# Solving a linear system
import numpy as np

# Coefficient matrix
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3,  1, -1]])

# Constants vector
B = np.array([[-1],
              [-3],
              [9]])

# Solve using inverse
A_inv = np.linalg.inv(A)
x = np.dot(A_inv, B)

print("Solution x:\n", x)
```

```python
# Experiment: Comparing Python lists vs NumPy arrays
import time
import numpy as np

# Create large lists
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))

# Element-wise addition using Python lists
start_time = time.time()
list_result = [list1[i] + list2[i] for i in range(1_000_000)]
end_time = time.time()
print("Time taken for element-wise addition using Python lists:",
      end_time - start_time, "seconds")

# Element-wise addition using NumPy arrays
arr1 = np.array(list1)
arr2 = np.array(list2)

start_time = time.time()
arr_result = arr1 + arr2
end_time = time.time()
print("Time taken using NumPy arrays:",
      end_time - start_time, "seconds")
```

```python
# Element-wise multiplication: Python lists vs NumPy arrays
import time
import numpy as np

# Create large lists
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))

# Element-wise multiplication using Python lists
start_time = time.time()
list_result = [list1[i] * list2[i] for i in range(1_000_000)]
end_time = time.time()
print("Time taken for element-wise multiplication using Python lists:",
      end_time - start_time, "seconds")

# Element-wise multiplication using NumPy arrays
arr1 = np.array(list1)
arr2 = np.array(list2)

start_time = time.time()
arr_result = arr1 * arr2
end_time = time.time()
print("Time taken using NumPy arrays:",
      end_time - start_time, "seconds")
```

```python
# Dot product: Python lists vs NumPy arrays
import time
import numpy as np

# Create large lists
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))

# Dot product using Python lists
start_time = time.time()
dot_product_list = 0
for i in range(1_000_000):
    dot_product_list += list1[i] * list2[i]
end_time = time.time()

print("Dot product using Python lists:", dot_product_list)
print("Time taken (Python lists):", end_time - start_time, "seconds")

# Dot product using NumPy arrays
arr1 = np.array(list1)
arr2 = np.array(list2)

start_time = time.time()
dot_product_numpy = np.dot(arr1, arr2)
end_time = time.time()

print("\nDot product using NumPy:", dot_product_numpy)
print("Time taken (NumPy):", end_time - start_time, "seconds")
```

```python
# Matrix multiplication: Python lists vs NumPy arrays
import time
import numpy as np
import random

# Matrix size
size = 1000

# Create two random matrices using Python lists
matrix1 = [[random.randint(0, 9) for _ in range(size)] for _ in range(size)]
matrix2 = [[random.randint(0, 9) for _ in range(size)] for _ in range(size)]

# Matrix multiplication using Python lists (manual)
start_time = time.time()

result_list = [[0]*size for _ in range(size)]
for i in range(size):
    for j in range(size):
        sum_val = 0
        for k in range(size):
            sum_val += matrix1[i][k] * matrix2[k][j]
        result_list[i][j] = sum_val

end_time = time.time()
print("Time taken for matrix multiplication using Python lists:", end_time - start_time, "seconds")

# Matrix multiplication using NumPy arrays
arr1 = np.array(matrix1)
arr2 = np.array(matrix2)

start_time = time.time()
result_numpy = np.dot(arr1, arr2)
end_time = time.time()
print("Time taken for matrix multiplication using NumPy arrays:", end_time - start_time, "seconds")
```
